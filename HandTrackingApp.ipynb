```python
import tensorflow as tf
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2, preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
import os

# Check your current working directory
print("Current working directory:", os.getcwd())

# Load pre-trained MobileNetV2 model
model = MobileNetV2(weights='imagenet')

# Replace with your actual image path
img_path = "Dataset/animals/ea36b90c2bf0083ed1584d05fb1d4e9fe777ead218ac104497f5c978a7eebdbb_640.jpg"

# Verify if file exists to avoid FileNotFoundError
if not os.path.exists(img_path):
    raise FileNotFoundError(f"Image file not found at path: {img_path}")

# Load and preprocess the image
img = image.load_img(img_path, target_size=(224, 224))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array = preprocess_input(img_array)

# Predict the class of the image
predictions = model.predict(img_array)

# Decode the predictions into class names
decoded_predictions = decode_predictions(predictions, top=3)[0]

# Display the image
plt.imshow(img)
plt.axis('off')
plt.show()

# Display the top 3 predictions with their scores
for i, (imagenet_id, label, score) in enumerate(decoded_predictions):
    print(f"{i + 1}. {label}: {score * 100:.2f}%")
```
